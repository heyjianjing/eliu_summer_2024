#### Some preparation first...

##### `Euler-Lagrange equation`

Let's say we want to find a function $x(t):[a, b] \rightarrow \mathbf{R}$ to minimize

$$G(x)=\int_a^b f\left(t, x(t), x'(t)\right)dt$$

Recall in calculus, the idea of minimizing a function is to look for its stationary points, where the `derivative` is zero

In order to find derivative of $G(x)$, we can `perturb` its input by just a bit and see what its output is doing, here, we want to perturb $x$, which itself is a `function`

To do this, assume there is another `arbitrary` function $\alpha(t):[a, b] \rightarrow \mathbf{R}$, with $\alpha(a)=\alpha(b)=0$

Then, we can consider

$$x_{\epsilon}(t)=x(t)+\epsilon \alpha(t)$$

as the `family of perturbed trajectories` of $x$ parameterized by $\epsilon$, and $\boxed{x_o(t)=x(t)}$

With this setup, the original objective can be rephrased as

`If` $x_{\epsilon}|_{\epsilon=0}:[a, b] \rightarrow \mathbf{R}$ minimizes the following function

$$G(\epsilon)=\int_a^bf\left(t, x_{\epsilon}(t), x_{\epsilon}'(t)\right) dt$$

`then` (drop $(t)$ to make it more readable)

$$\left.\frac{dG}{d\epsilon}\right|_{\epsilon=0}=\boxed{\left.\frac{d}{d\epsilon}\right|_{\epsilon=0}\int_a^bf\left(t, x_{\epsilon}, x_{\epsilon}'\right) dt=0}$$

Move the derivative into the integral

$$\begin{align*}
\int_a^b \left.\frac{\partial}{\partial \epsilon} f\left(t, x_{\epsilon}, x_{\epsilon}'\right)\right|_{\epsilon=0} dt&=\int_a^b \left.\left[\frac{\partial f}{\partial x_{\epsilon}}\frac{\partial x_{\epsilon}}{\partial \epsilon}+\frac{\partial f}{\partial x_{\epsilon}'}\frac{\partial x_{\epsilon}'}{\partial \epsilon}\right]\right|_{\epsilon=0} dt\\
&=\int_a^b \left.\left[\frac{\partial f}{\partial x_{\epsilon}}\alpha+\frac{\partial f}{\partial x_{\epsilon}'}\alpha'\right]\right|_{\epsilon=0} dt\\
& u=\frac{\partial f}{\partial x_{\epsilon}'},\, dv=\alpha' dt \rightarrow du=\frac{d}{dt}\left[\frac{\partial f}{\partial x_{\epsilon}'}\right] dt\, ,v=\alpha\\
&=\int_a^b \left.\left(\frac{\partial f}{\partial x_{\epsilon}}\alpha \, dt
+ \left.\frac{\partial f}{\partial x_{\epsilon}'}\alpha\right|_a^b - \int_a^b \alpha \frac{d}{dt}\left[\frac{\partial f}{\partial x_{\epsilon}'}\right] \right)\right|_{\epsilon=0}dt \\
& \alpha(a)=\alpha(b)=0 \rightarrow \left.\frac{\partial f}{\partial x_{\epsilon}'}\alpha\right|_a^b=0 \\
&=\int_a^b \left.\left(\frac{\partial f}{\partial x_{\epsilon}}\alpha - \alpha \frac{d}{dt}\left[\frac{\partial f}{\partial x_{\epsilon}'}\right] \right)\right|_{\epsilon=0}dt \\
&=0
\end{align*}$$

Since $\alpha(t)$ is arbitrary, implying that the integrand must be zero `pointwise` for the equality to hold, so we can simplify to get the `Euler-Lagrange equation`

$$\begin{align*}
0&= \int_a^b \left.\left(\frac{\partial f}{\partial x_{\epsilon}} -  \frac{d}{dt}\left[\frac{\partial f}{\partial x_{\epsilon}'}\right] \right)\right|_{\epsilon=0}dt\\
&= \left.\left(\frac{\partial f}{\partial x_{\epsilon}} -  \frac{d}{dt}\left[\frac{\partial f}{\partial x_{\epsilon}'}\right] \right)\right|_{\epsilon=0}\\
&=\boxed{\frac{\partial f}{\partial x} -  \frac{d}{dt}\left[\frac{\partial f}{\partial x'}\right] }
\end{align*}$$

Euler-Lagrange equation is a `necessary` condition for $G(x)$ to be `stationary`. However, it does not tell whether it is a maximum or minimum

##### `Lagrangian` mechanics

One thing for which we can use the Euler-Lagrange equation is the `principle of least action`

Intuitively, for example, for a physical particle to move around from $x(t_1)$ to $x(t_2)$, then, the principle of least action says that the actual trajectory (i.e., function) it takes would be a stationary point for

$$\int_{t_1}^{t_2} L\left(t, x(t), x'(t)\right)dt$$

where

$$L\left(t, x(t), x'(t)\right)=K-U=\frac{1}{2}mx'^2(x)-U(x(t))$$

is called `Lagrangian` of the particle, $K$ denotes `kinetic energy` and $U$ denotes generalized `potential energy`

As a result

$$\frac{\partial L}{\partial x} -  \frac{d}{dt}\left[\frac{\partial L}{\partial x'}\right]=0$$

If we plug in the Lagrangian above, we will see that the Euler-Lagrange equation becomes

$$-\frac{\partial U(x(t))}{\partial x}=mx''(t)$$

which is Newton's `second law of motion`

Therefore, Lagrangian mechanics provides us a new formulation of motion through `energy`

In addition, the Euler-Lagrange equation is agnostic to the `coordinate system`, so we will use `generalized position` $q_i$ just to follow the convention

$$\frac{\partial L}{\partial q_i} -  \frac{d}{dt}\left[\frac{\partial L}{\partial q_i'}\right]=0$$

##### `Hamiltonian` equations

Interestingly, while Lagrangian mechanics uses energy, the Lagrangian is actually the `difference` between kinetic and potential energy. But how about the `total energy` we are usually familiar with?

It turns out, the `total energy` of the system can be expressed as

$$\begin{align*}
H&= \frac{\partial L}{\partial q'}q'-L \\
&= mq'q'-(K-U)\\
&=2K-K+U\\
&=K+U
\end{align*}$$

Further, we see that $\frac{\partial L}{\partial q'}=mq'$, which can be considered `generalized momentum` corresponding to generalized position $q$

We denote it using a new variable $p$ such that $p=\frac{\partial L}{\partial q'}$, and we have the `Hamiltonian`

$$H(q, p, t)=pq'-L(q, q', t)$$

Take total differential on both side

$$\frac{\partial H}{\partial q}dq+\frac{\partial H}{\partial p}dp+\frac{\partial H}{\partial t}dt = q'dp - pdq'-\frac{\partial L}{\partial q}dq-\frac{\partial L}{\partial q'}dq'-\frac{\partial L}{\partial t}dt$$

Compare coefficients, we have the set of `Hamiltonian equations`

$$\begin{align*}
\frac{\partial H}{\partial q}&=-\frac{\partial L}{\partial q}=-\frac{d}{dt}\left[\frac{\partial L}{\partial q'}\right]=-p'\\
\frac{\partial H}{\partial p}&=q' \\
\frac{\partial H}{\partial t}&=-\frac{\partial L}{\partial t}
\end{align*}$$

An important property is that the Hamiltonian is conserved

$$\frac{\partial H}{\partial t}=\frac{\partial H}{\partial q}\frac{\partial q}{\partial t}+\frac{\partial H}{\partial p}\frac{\partial p}{\partial t}=-p'q'+q'p'=0\$$

This allows us to evolve $p$ and $q$ over $t$

##### `Leapfrog` method

One way we could do this is through leapfrog method that alternates the following equations in L-1 intermediate steps

$$\begin{align*}
p(t+\frac{\epsilon}{2})&=p(t)-\frac{\epsilon}{2}\frac{\partial H}{\partial q}(q(t)) \\
q(t+\epsilon)&=q(t)+\epsilon p(t+\frac{\epsilon}{2}) \\
\end{align*}$$

and in the last step after the final `position` update, the `momentum` is updated with a second half-step

$$\begin{align*}
p(t+\epsilon)&=p(t+\frac{\epsilon}{2})-\frac{\epsilon}{2}\frac{\partial H}{\partial q}(q(t+\epsilon))
\end{align*}$$

##### `Hamiltonian Monte Carlo` sampling (HMC)

Finally, we can get to HMC

Commonly, system of N particles from thermodynamics is used given by Boltzman distribution

$$prob_j\propto e^{-\frac{E_j}{kT}}$$

where $prob_j$ is the probability of the system being in state $j$, $E_j$ is the energy of state $j$, and $k$ is Boltzman constant

In HMC, we draw an analogy between thermodynamics and probability distributions. Specifically, we let the `potential` energy be

$$U(q)=-\log f(q)$$

where $f(q)$ is the distribution we want for `positions`, and `kinetic` energy as (with $p_i$ being the `momentum` associated with the ith particle)

$$K(p)=\sum_i^N\frac{1}{2m_i}p_i^2$$

Then, the `joint probabilty distribution` (set $kT=1$ and $m_i=1$) is `decoupled`

$$\begin{align*}
prob(q,p)&=\frac{1}{Z}\exp \left(-\frac{H(q, p)}{kT}\right) \\
&=\frac{1}{Z}\exp\left(\log f(q)-\sum_{i=1}^N\frac{1}{2}p_i^2\right) \\
&=\frac{1}{Z_1}f(q)\cdot\frac{1}{Z_2}\exp\left(-\sum_{i=1}^N\frac{1}{2}p_i^2\right)
\end{align*}$$

Now, we use the leapfrog algorihtm to evolve the system to get the proposed state $q^*, p^*$

After this, we negate the momentum variable $p^*$ (which does not affect the Hamiltonian because the momentum is squared in the kinetic energy term). This negation is required for the proof of HMC

With the proposed state, we then follow the Metropolis algorithm to accept the new state with probabilty

$$\frac{prob(q^*, p^*)}{prob(q, p)}=\frac{\exp\left(-H(q^*, p^*)\right)}{\exp\left(-H(q, p)\right)}=\exp\left(H(q, p)-H(q^*, p^*)\right)$$

#### Example

We again consider the following expression

$$f(x) = \exp(-x^{2})(2 + \sin(3x) + \sin(4x)), \quad \text{for } x \in [0, 3]$$

and normalize it to turn it into a valid probability distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from scipy.integrate import quad
import time
plt.style.use('dark_background')
np.set_printoptions(formatter={'float': '{: 0.4f}'.format})
```

```python
# Unnormalized distribution
def target_pdf(x):
    # Ensure that x is non-negative
    return np.exp(-x**2) * (2 + np.sin(3 * x) + np.sin(4 * x))

# Compute normalizing factor using numerical integration
normalizing_factor, _ = quad(target_pdf, 0, 3)

# Normalized distribution
def normalized_pdf(x):
    return target_pdf(x) / normalizing_factor
```

```python
x = np.linspace(0, 3, 1000)
y_unnormalized = target_pdf(x)
y_normalized = normalized_pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_unnormalized, label='Unnormalized PDF', linewidth=2)
plt.plot(x, y_normalized, label='Normalized PDF', linewidth=2, linestyle='--')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.title('Unnormalized and normalized PDF from an arbitrary function $f(x)$')
plt.show()
```

```python
def potential_energy(q):
    return -np.log(target_pdf(q))

def grad_potential_energy(q):
    epsilon = 1e-5
    return (potential_energy(q + epsilon) - potential_energy(q - epsilon)) / (2 * epsilon)

def leapfrog(q, p, step_size, num_steps):
    p = p - 0.5 * step_size * grad_potential_energy(q)  # Half-step update of momentum
    for i in range(num_steps):
        q = q + step_size * p  # Full-step update of position
        if i != num_steps - 1:
            p = p - step_size * grad_potential_energy(q)  # Full-step update of momentum
    p = p - 0.5 * step_size * grad_potential_energy(q)  # Final half-step update of momentum
    return q, p

def hamiltonian(q, p):
    return potential_energy(q) + 0.5 * np.sum(p**2)

def hmc_sampling(num_samples, initial_q, step_size, num_steps):
    samples = []
    q = initial_q
    accepted = 0
    for _ in range(num_samples):
        p = np.random.normal(0, 1, size=q.shape)  # Fresh new random momentum
        initial_h = hamiltonian(q, p)

        new_q, new_p = leapfrog(q, p, step_size, num_steps)

        if 0 <= new_q <= 3:
            new_h = hamiltonian(new_q, new_p)

            if np.random.rand() < np.exp(initial_h - new_h):
                q = new_q # This becomes initial q for next sample, also the required property of Markov chain
                accepted += 1
        samples.append(q)
    print(f'Accepted: {accepted}')
    print(f'Acceptance rate: {accepted / num_samples:.4f}')
    return np.array(samples)
```

```python
initial_q = np.array([1.0])  # Initial position
step_size = 0.1  # Step size for leapfrog
num_steps = 10  # Number of leapfrog steps
num_samples =  2000  # Number of samples

start_time = time.time()
samples = hmc_sampling(num_samples, initial_q, step_size, num_steps)
print(f'Time taken: {time.time() - start_time:.4f} seconds')

x = np.linspace(0, 3, 1000)
y_unnormalized = target_pdf(x)
y_normalized = normalized_pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_unnormalized, label='Unnormalized PDF', linewidth=2)
plt.plot(x, y_normalized, label='Normalized PDF', linewidth=2, linestyle='--')
plt.hist(samples, bins=40, density=True, alpha=0.6, label='HMC samples', edgecolor='black')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.title('Hamiltonian Monte Carlo Sampling')
plt.show()
```
